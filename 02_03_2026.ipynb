```python
# ==========================================
# Penguin Dashboard using Plotly & Dash
# ==========================================

import numpy as np
import pandas as pd
import seaborn as sns

import plotly.express as px
import plotly.graph_objects as go

from dash import Dash, dcc, html
from dash.dependencies import Input, Output


# ==========================================
# Load and Prepare Data
# ==========================================

df = sns.load_dataset("penguins")

# Clean data
df = df.dropna()

# Create new feature
df["bill_ratio"] = df["bill_length_mm"] / df["bill_depth_mm"]


# ==========================================
# KPIs
# ==========================================

total_penguins = len(df)

avg_body_mass = round(df["body_mass_g"].mean(), 2)

avg_flipper = round(df["flipper_length_mm"].mean(), 2)

avg_bill_length = round(df["bill_length_mm"].mean(), 2)


# ==========================================
# Create App
# ==========================================

app = Dash(__name__)


# ==========================================
# Layout
# ==========================================

app.layout = html.Div([

    html.H1("Penguin Analysis Dashboard",
            style={'textAlign': 'center'}),

    # KPIs
    html.Div([

        html.Div([
            html.H3("Total Penguins"),
            html.H2(total_penguins)
        ], className="card"),

        html.Div([
            html.H3("Avg Body Mass"),
            html.H2(avg_body_mass)
        ], className="card"),

        html.Div([
            html.H3("Avg Flipper Length"),
            html.H2(avg_flipper)
        ], className="card"),

        html.Div([
            html.H3("Avg Bill Length"),
            html.H2(avg_bill_length)
        ], className="card"),

    ], style={"display": "flex", "justifyContent": "space-around"}),


    # Dropdown Filter
    html.Div([

        html.Label("Select Species"),

        dcc.Dropdown(

            options=[

                {"label": "All", "value": "All"},
                {"label": "Adelie", "value": "Adelie"},
                {"label": "Gentoo", "value": "Gentoo"},
                {"label": "Chinstrap", "value": "Chinstrap"},

            ],

            value="All",

            id="species_filter"

        )

    ], style={"width": "40%"}),


    # Graphs
    dcc.Graph(id="species_chart"),

    dcc.Graph(id="scatter_chart"),

    dcc.Graph(id="box_chart"),

    dcc.Graph(id="correlation_chart")


])


# ==========================================
# Callback
# ==========================================

@app.callback(

    Output("species_chart", "figure"),
    Output("scatter_chart", "figure"),
    Output("box_chart", "figure"),
    Output("correlation_chart", "figure"),

    Input("species_filter", "value")

)

def update_dashboard(selected_species):

    if selected_species == "All":

        filtered_df = df

    else:

        filtered_df = df[df["species"] == selected_species]


    # Bar Chart

    fig1 = px.bar(

        filtered_df,

        x="species",

        title="Species Count",

        color="species"

    )


    # Scatter Chart

    fig2 = px.scatter(

        filtered_df,

        x="flipper_length_mm",

        y="body_mass_g",

        color="species",

        title="Flipper Length vs Body Mass"

    )


    # Box Plot

    fig3 = px.box(

        filtered_df,

        x="species",

        y="body_mass_g",

        title="Body Mass Distribution"

    )


    # Correlation Heatmap

    corr = filtered_df.corr(numeric_only=True)

    fig4 = px.imshow(

        corr,

        text_auto=True,

        title="Correlation Heatmap"

    )


    return fig1, fig2, fig3, fig4


# ==========================================
# Run App
# ==========================================

if __name__ == "__main__":

    app.run(debug=True)
```
