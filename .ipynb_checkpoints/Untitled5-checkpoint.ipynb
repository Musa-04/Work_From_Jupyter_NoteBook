```python
import pandas as pd

data = {
    "Name": ["Ali", "John", "Sara", "David"],
    "Math": [85, 90, 78, 92],
    "Science": [88, 76, 90, 85]
}

df = pd.DataFrame(data)

print("Original Data")
print(df)


# Add total

df["Total"] = df["Math"] + df["Science"]

print("\nWith Total")
print(df)


# Average

df["Average"] = df["Total"] / 2

print("\nWith Average")
print(df)


# Top student

print("\nTop Student")

print(df[df["Average"] == df["Average"].max()])
```

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("company.db")
```

```python
conn.execute("""
CREATE TABLE IF NOT EXISTS employees(
id INTEGER PRIMARY KEY,
name TEXT,
salary INTEGER,
department TEXT
)
""")
```

```python
data = {
    "id": [1, 2, 3],
    "name": ["Abhi", "Jay", "Amit"],
    "salary": [25000, 30000, 28000],
    "department": ["IT", "HR", "Finance"]
}

df = pd.DataFrame(data)

df.to_sql("employees", conn, if_exists="append", index=False)

print("Data Inserted")
```

```python
df = pd.read_sql("SELECT * FROM employees", conn)

print(df)
```

```python
conn.execute("""
UPDATE employees
SET salary = 40000
WHERE name = 'Abhi'
""")

conn.commit()

print("Updated")
```

```python
conn.execute("""
DELETE FROM employees
WHERE name = 'Jay'
""")

conn.commit()

print("Deleted")
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
import sqlite3
import pandas as pd
import datetime


# ---------------- DATABASE ---------------- #

conn = sqlite3.connect("hotel_pos.db")


# Users

conn.execute("""

CREATE TABLE IF NOT EXISTS users(

username TEXT PRIMARY KEY,
password TEXT

)

""")

conn.execute("""

INSERT OR IGNORE INTO users VALUES(

'admin','admin123'

)

""")


# Menu

conn.execute("""

CREATE TABLE IF NOT EXISTS menu(

id INTEGER PRIMARY KEY AUTOINCREMENT,
name TEXT,
price REAL,
cost REAL

)

""")


# Orders

conn.execute("""

CREATE TABLE IF NOT EXISTS orders(

date TEXT,
table_no INTEGER,
item TEXT,
qty INTEGER,
price REAL,
total REAL,
profit REAL

)

""")


conn.commit()


# Default Menu

menu_data = [

("Biryani",200,120),

("Fried Rice",150,90),

("Dosa",80,40),

("Tea",20,5),

("Coffee",30,10),

("Burger",120,70)

]

for item in menu_data:

    conn.execute("""

    INSERT OR IGNORE INTO menu(name,price,cost)

    VALUES(?,?,?)

    """,item)

conn.commit()


# ---------------- LOGIN ---------------- #

def login():

    user = user_entry.get()

    pwd = pass_entry.get()


    data = conn.execute(

    "SELECT * FROM users WHERE username=? AND password=?",

    (user,pwd)).fetchall()


    if data:

        login_win.destroy()

        dashboard()

    else:

        messagebox.showerror("Error","Invalid Login")



# ---------------- DASHBOARD ---------------- #

def dashboard():

    dash = tk.Tk()

    dash.title("Hotel POS")

    dash.geometry("900x600")

    dash.configure(bg="#34495e")


    tk.Label(dash,

    text="Hotel POS System",

    font=("Arial",22),

    fg="white",

    bg="#34495e").pack(pady=20)


    tk.Button(dash,

    text="New Order",

    width=20,

    command=new_order).pack(pady=10)


    tk.Button(dash,

    text="Sales Report",

    width=20,

    command=sales_report).pack(pady=10)


    tk.Button(dash,

    text="Profit Report",

    width=20,

    command=profit_report).pack(pady=10)



    dash.mainloop()



# ---------------- NEW ORDER ---------------- #

def new_order():

    order_win = tk.Toplevel()

    order_win.title("Order")

    order_win.geometry("700x500")


    # Table number

    tk.Label(order_win,

    text="Table No").pack()


    table_combo = ttk.Combobox(order_win,

    values=list(range(1,21)))

    table_combo.pack()


    # Menu

    menu = pd.read_sql(

    "SELECT name FROM menu",conn)


    tk.Label(order_win,

    text="Menu Item").pack()


    item_combo = ttk.Combobox(

    order_win,

    values=list(menu["name"]))

    item_combo.pack()


    tk.Label(order_win,

    text="Quantity").pack()


    qty_entry = tk.Entry(order_win)

    qty_entry.pack()


    tree = ttk.Treeview(order_win)

    tree["columns"]=("Item","Qty","Price","Total")

    for col in tree["columns"]:

        tree.heading(col,text=col)

    tree.pack()


    def add():

        item = item_combo.get()

        qty = int(qty_entry.get())


        data = conn.execute("""

        SELECT price,cost FROM menu

        WHERE name=?

        """,(item,)).fetchone()


        price, cost = data


        total = price * qty


        tree.insert("",tk.END,

        values=(item,qty,price,total))



    def save():

        table_no = table_combo.get()

        date = str(datetime.date.today())


        for row in tree.get_children():

            data = tree.item(row)["values"]

            item, qty, price, total = data


            cost = conn.execute("""

            SELECT cost FROM menu

            WHERE name=?

            """,(item,)).fetchone()[0]


            profit = (price-cost)*qty


            conn.execute("""

            INSERT INTO orders

            VALUES(?,?,?,?,?,?,?)

            """,(date,table_no,

            item,qty,price,total,profit))


        conn.commit()

        messagebox.showinfo("Saved","Order Saved")



    tk.Button(order_win,

    text="Add Item",

    command=add).pack()


    tk.Button(order_win,

    text="Save Bill",

    command=save).pack()



# ---------------- SALES REPORT ---------------- #

def sales_report():

    win = tk.Toplevel()

    df = pd.read_sql(

    "SELECT * FROM orders",conn)


    text = tk.Text(win)

    text.pack()

    text.insert(tk.END,str(df))



# ---------------- PROFIT ---------------- #

def profit_report():

    win = tk.Toplevel()

    df = pd.read_sql(

    "SELECT SUM(profit) FROM orders",

    conn)


    tk.Label(win,

    text="Total Profit",

    font=("Arial",20)).pack()


    tk.Label(win,

    text=str(df.iloc[0,0]),

    font=("Arial",20)).pack()



# ---------------- LOGIN WINDOW ---------------- #

login_win = tk.Tk()

login_win.title("Hotel Login")

login_win.geometry("400x300")


tk.Label(login_win,

text="Username").pack()

user_entry = tk.Entry(login_win)

user_entry.pack()


tk.Label(login_win,

text="Password").pack()

pass_entry = tk.Entry(login_win,

show="*")

pass_entry.pack()


tk.Button(login_win,

text="Login",

command=login).pack()


login_win.mainloop()
```
