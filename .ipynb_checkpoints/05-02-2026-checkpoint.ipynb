```python
# Program to count the number of errors
# and store the error lines separately

log_lines = [
    "INFO: Application started",
    "ERROR: Failed to connect to database",
    "WARNING: Low memory",
    "ERROR: Invalid user input",
    "INFO: Process completed"
]

error_count = 0
error_lines = []

for line in log_lines:
    if "ERROR" in line:
        error_count += 1
        error_lines.append(line)

print("Total number of errors:", error_count)
print("Error lines:")
for error in error_lines:
    print(error)
```

```python
# Apply file handling + exception handling
# Use Case: Store student marks safely in a file

try:
    # Take input from user
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))

    # Validate marks
    if marks < 0 or marks > 100:
        raise ValueError("Marks must be between 0 and 100")

    # File handling (write mode)
    file = open("student_marks.txt", "a")
    file.write(f"{name} - {marks}\n")
    file.close()

    print("Student marks saved successfully")

except ValueError as ve:
    print("Value Error:", ve)

except FileNotFoundError:
    print("File not found")

except Exception as e:
    print("Unexpected error:", e)
```

```python
# ATM Mini Project (Clean & User-Friendly for Jupyter)

balance = 10000
correct_pin = "1234"

print("üèß Welcome to the ATM System")

pin = input("üîê Enter your ATM PIN: ")

if pin != correct_pin:
    print("‚ùå Incorrect PIN. Access Denied.")
else:
    print("‚úÖ Login Successful")

    while True:
        print("\n==========================")
        print("üìã ATM MENU")
        print("1Ô∏è‚É£  Check Balance")
        print("2Ô∏è‚É£  Deposit Money")
        print("3Ô∏è‚É£  Withdraw Money")
        print("4Ô∏è‚É£  Exit")
        print("==========================")

        choice = input("üëâ Enter your choice (1-4): ")

        if choice == "1":
            print(f"üí∞ Current Balance: ‚Çπ{balance}")

        elif choice == "2":
            amount = float(input("‚ûï Enter deposit amount: ‚Çπ"))
            if amount > 0:
                balance += amount
                print(f"‚úÖ Deposit Successful! New Balance: ‚Çπ{balance}")
            else:
                print("‚ö†Ô∏è Invalid amount. Please enter a positive value.")

        elif choice == "3":
            amount = float(input("‚ûñ Enter withdrawal amount: ‚Çπ"))
            if 0 < amount <= balance:
                balance -= amount
                print("üíµ Please collect your cash.")
                print(f"üìâ Remaining Balance: ‚Çπ{balance}")
            else:
                print("‚ö†Ô∏è Insufficient balance or invalid amount.")

        elif choice == "4":
            print("üôè Thank you for using the ATM. Have a nice day!")
            break

        else:
            print("‚ö†Ô∏è Invalid choice. Please select between 1 and 4.")
```

```python
# Student Record Management System
# Clean & User-Friendly (Jupyter Notebook)

students = []   # list to store student records

print("üéì Welcome to Student Record Management System")

while True:
    print("\n==============================")
    print("üìã MENU")
    print("1Ô∏è‚É£  Add Student Record")
    print("2Ô∏è‚É£  View All Student Records")
    print("3Ô∏è‚É£  Search Student by Roll No")
    print("4Ô∏è‚É£  Delete Student Record")
    print("5Ô∏è‚É£  Exit")
    print("==============================")

    choice = input("üëâ Enter your choice (1-5): ")

    # -------- ADD STUDENT --------
    if choice == "1":
        roll = input("Enter Roll Number: ")
        name = input("Enter Student Name: ")
        marks = float(input("Enter Marks: "))

        student = {
            "roll": roll,
            "name": name,
            "marks": marks
        }

        students.append(student)
        print("‚úÖ Student record added successfully")

    # -------- VIEW ALL STUDENTS --------
    elif choice == "2":
        if not students:
            print("‚ö†Ô∏è No student records found")
        else:
            print("\nüìö Student Records:")
            for s in students:
                print(f"Roll: {s['roll']} | Name: {s['name']} | Marks: {s['marks']}")

    # -------- SEARCH STUDENT --------
    elif choice == "3":
        search_roll = input("Enter Roll Number to search: ")
        found = False

        for s in students:
            if s["roll"] == search_roll:
                print("üîç Student Found")
                print(f"Roll: {s['roll']} | Name: {s['name']} | Marks: {s['marks']}")
                found = True
                break

        if not found:
            print("‚ùå Student not found")

    elif choice == "4":
        delete_roll = input("Enter Roll Number to delete: ")
        for s in students:
            if s["roll"] == delete_roll:
                students.remove(s)
                print("üóëÔ∏è Student record deleted successfully")
                break
        else:
            print("‚ùå Student not found")

    elif choice == "5":
        print("üôè Exiting Student Record Management System")
        break

    else:
        print("‚ö†Ô∏è Invalid choice. Please enter between 1 and 5.")
```
