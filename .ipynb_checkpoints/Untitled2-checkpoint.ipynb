```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg


class Dashboard:

    def __init__(self, root):

        self.root = root
        self.root.title("Thinker Data Dashboard")
        self.root.geometry("1100x700")

        self.df = None

        title = tk.Label(root, text="Thinker Data Visualization Dashboard",
                         font=("Arial", 20, "bold"))
        title.pack(pady=10)

        # Buttons
        frame = tk.Frame(root)
        frame.pack()

        ttk.Button(frame, text="Upload CSV/Excel",
                   command=self.upload).grid(row=0, column=0, padx=10)

        ttk.Button(frame, text="Scatter",
                   command=self.scatter).grid(row=0, column=1)

        ttk.Button(frame, text="Bar",
                   command=self.bar).grid(row=0, column=2)

        ttk.Button(frame, text="Histogram",
                   command=self.hist).grid(row=0, column=3)

        ttk.Button(frame, text="Box",
                   command=self.box).grid(row=0, column=4)

        ttk.Button(frame, text="Heatmap",
                   command=self.heatmap).grid(row=0, column=5)

        # Chart area
        self.chart = tk.Frame(root)
        self.chart.pack(fill="both", expand=True)


    def upload(self):

        file = filedialog.askopenfilename(
            filetypes=[("CSV files", "*.csv"),
                       ("Excel files", "*.xlsx")]
        )

        if file:

            if file.endswith(".csv"):
                self.df = pd.read_csv(file)

            else:
                self.df = pd.read_excel(file)

            messagebox.showinfo("Success", "Data Loaded Successfully")


    def clear(self):

        for i in self.chart.winfo_children():
            i.destroy()


    def show(self, fig):

        canvas = FigureCanvasTkAgg(fig, self.chart)
        canvas.draw()
        canvas.get_tk_widget().pack(fill="both", expand=True)


    def scatter(self):

        if self.df is None:
            return

        self.clear()

        num = self.df.select_dtypes(include="number").columns

        fig, ax = plt.subplots()

        sns.scatterplot(data=self.df,
                        x=num[0],
                        y=num[1],
                        palette="viridis",
                        ax=ax)

        self.show(fig)


    def bar(self):

        if self.df is None:
            return

        self.clear()

        cat = self.df.select_dtypes(include="object").columns
        num = self.df.select_dtypes(include="number").columns

        fig, ax = plt.subplots()

        sns.barplot(data=self.df,
                    x=cat[0],
                    y=num[0],
                    palette="magma",
                    ax=ax)

        self.show(fig)


    def hist(self):

        if self.df is None:
            return

        self.clear()

        num = self.df.select_dtypes(include="number").columns

        fig, ax = plt.subplots()

        sns.histplot(self.df[num[0]],
                     kde=True,
                     color="purple",
                     ax=ax)

        self.show(fig)


    def box(self):

        if self.df is None:
            return

        self.clear()

        cat = self.df.select_dtypes(include="object").columns
        num = self.df.select_dtypes(include="number").columns

        fig, ax = plt.subplots()

        sns.boxplot(data=self.df,
                    x=cat[0],
                    y=num[0],
                    palette="coolwarm",
                    ax=ax)

        self.show(fig)


    def heatmap(self):

        if self.df is None:
            return

        self.clear()

        fig, ax = plt.subplots()

        sns.heatmap(self.df.corr(),
                    cmap="inferno",
                    annot=True,
                    ax=ax)

        self.show(fig)


root = tk.Tk()

Dashboard(root)

root.mainloop()
```
