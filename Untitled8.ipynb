```python
import pandas as pd
import numpy as np
n=500
np.random.seed(42)
data=pd.DataFrame({
    "Age":np.random.randint(18,60,n),
    "Income":np.random.normal(50000,15000,n),
    "Spending_Score":np.random.randint(1,100,n),
    "Gender":np.random.choice(["male","female"],n),
    "City":np.random.choice(["Hyderabad","delhi","Mumbai"],n),
    "Purchase_Amount":np.random.normal(5000,2000,n),
    "Customer_rating":np.random.uniform(1,5,n)
})
```

```python
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(data=data.sort_values("Age"),
             x="Age",
y="Purchase_Amount")
plt.title("purchase amount vs age")
```

```python
sns.scatterplot(data=data, x="Income",y="Spending_Score",hue="Gender")
```

```python
sns.barplot(data=data,x="City",y="Purchase_Amount")
```

```python
sns.boxplot(data=data,x="Gender" , y="Purchase_Amount")
plt.show()
```

```python
sns.violinplot(data=data,x="Gender",y="Spending_Score")
plt.title("Spending Score distribution")
```

```python
sns.jointplot(data=data)
```

```python
sns.pairplot(data,hue="Gender")
plt.
```
