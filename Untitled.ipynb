```python
#task1
#employee Salary Calculator 

base_salary = float(input("Enter the Base Salary: "))
bonus = float(input("Enter the Bonus: "))
tax_percent =float(input("Enter the Tax Percentage: "))

total_salary = base_salary + bonus
tax_amount = (tax_percent*total_salary) / 100
final_salary=total_salary-tax_amount

print("\n--- Salary Details---")
print("total_Salary:", total_salary)
print("Tax Amount:", tax_amount)
print("Final Monthly Salary:", final_salary)
```
