```python
import streamlit as st
import pandas as pd
import os

FILE_NAME = "grades.txt"


# ---------------------------
# Grade Logic
# ---------------------------
def calculate_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= 40:
        return "D"
    else:
        return "F"


# ---------------------------
# Save Student
# ---------------------------
def save_student(name, m1, m2, m3):
    total = m1 + m2 + m3
    avg = round(total / 3, 2)
    grade = calculate_grade(avg)

    with open(FILE_NAME, "a") as f:
        f.write(f"{name},{m1},{m2},{m3},{total},{avg},{grade}\n")


# ---------------------------
# Load Students
# ---------------------------
def load_students():
    if not os.path.exists(FILE_NAME):
        return pd.DataFrame(
            columns=["Name", "M1", "M2", "M3", "Total", "Average", "Grade"]
        )

    return pd.read_csv(
        FILE_NAME,
        names=["Name", "M1", "M2", "M3", "Total", "Average", "Grade"]
    )


# ---------------------------
# UI
# ---------------------------
st.title("ðŸŽ“ Student Grade Book Manager")

menu = st.sidebar.selectbox(
    "Menu",
    ["Add Student", "View Students"]
)


# ---------------------------
# Add Student Page
# ---------------------------
if menu == "Add Student":

    st.header("Add New Student")

    name = st.text_input("Student Name")

    col1, col2, col3 = st.columns(3)

    with col1:
        m1 = st.number_input("Mark 1", 0, 100, 0)

    with col2:
        m2 = st.number_input("Mark 2", 0, 100, 0)

    with col3:
        m3 = st.number_input("Mark 3", 0, 100, 0)

    if st.button("Save Student"):
        if name.strip() == "":
            st.error("Name cannot be empty")
        else:
            save_student(name, m1, m2, m3)
            st.success("Student added successfully âœ…")


# ---------------------------
# View Students Page
# ---------------------------
elif menu == "View Students":

    st.header("All Students")

    df = load_students()

    if df.empty:
        st.warning("No records found")
    else:
        st.dataframe(df, use_container_width=True)

        st.subheader("Statistics")
        st.bar_chart(df.set_index("Name")["Average"])
```
