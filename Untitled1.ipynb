```python
import matplotlib.pyplot as plt

ypoints = [3, 7, 2, 9, 5]

plt.plot(ypoints, marker='p')

plt.title("Pentagon Marker Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()
```

```python
plt.plot(ypoints, marker='p', linestyle='--', color='red', markersize=10)
plt.show()
```
