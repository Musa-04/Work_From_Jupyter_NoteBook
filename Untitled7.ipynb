```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import pandas as pd
import matplotlib.pyplot as plt


class DataApp:

    def __init__(self, root):

        self.root = root
        self.root.title("Advanced Data Management System")
        self.root.geometry("1300x700")

        self.df = None

        self.create_ui()


    def create_ui(self):

        title = tk.Label(
            self.root,
            text="Data Management Dashboard",
            font=("Arial", 20, "bold"),
            bg="#2c3e50",
            fg="white"
        )

        title.pack(fill="x")


        frame = tk.Frame(self.root)
        frame.pack(fill="x")


        # File Button
        tk.Button(frame, text="Upload File", command=self.upload_file).pack(side="left")


        # Chart Buttons
        tk.Button(frame, text="Line", command=self.line_chart).pack(side="left")

        tk.Button(frame, text="Bar", command=self.bar_chart).pack(side="left")

        tk.Button(frame, text="Histogram", command=self.hist_chart).pack(side="left")


        # Pandas Operations
        tk.Button(frame, text="Head", command=self.show_head).pack(side="left")

        tk.Button(frame, text="Tail", command=self.show_tail).pack(side="left")

        tk.Button(frame, text="Shape", command=self.show_shape).pack(side="left")

        tk.Button(frame, text="Columns", command=self.show_columns).pack(side="left")

        tk.Button(frame, text="Info", command=self.show_info).pack(side="left")

        tk.Button(frame, text="Describe", command=self.show_describe).pack(side="left")

        tk.Button(frame, text="Values", command=self.show_values).pack(side="left")

        tk.Button(frame, text="Keys", command=self.show_keys).pack(side="left")

        tk.Button(frame, text="Dtypes", command=self.show_dtypes).pack(side="left")


        # Column select
        self.column_box = ttk.Combobox(frame)

        self.column_box.pack(side="left", padx=10)


        # Table
        self.table = ttk.Treeview(self.root)

        self.table.pack(fill="both", expand=True)


    # Upload File
    def upload_file(self):

        file = filedialog.askopenfilename(
            filetypes=[("CSV", "*.csv"), ("Excel", "*.xlsx")]
        )

        if file:

            if file.endswith(".csv"):

                self.df = pd.read_csv(file)

            else:

                self.df = pd.read_excel(file)


            self.show_table(self.df)

            self.column_box["values"] = list(self.df.columns)


    # Show table
    def show_table(self, data):

        self.table.delete(*self.table.get_children())

        self.table["column"] = list(data.columns)

        self.table["show"] = "headings"

        for col in data.columns:

            self.table.heading(col, text=col)


        for row in data.values:

            self.table.insert("", "end", values=row)


    # Pandas Functions

    def show_head(self):

        self.show_table(self.df.head())


    def show_tail(self):

        self.show_table(self.df.tail())


    def show_shape(self):

        messagebox.showinfo("Shape", str(self.df.shape))


    def show_columns(self):

        messagebox.showinfo("Columns", str(self.df.columns))


    def show_info(self):

        messagebox.showinfo("Info", str(self.df.info()))


    def show_describe(self):

        self.show_table(self.df.describe())


    def show_values(self):

        messagebox.showinfo("Values", str(self.df.values))


    def show_keys(self):

        messagebox.showinfo("Keys", str(self.df.keys()))


    def show_dtypes(self):

        messagebox.showinfo("Dtypes", str(self.df.dtypes))


    # Charts

    def line_chart(self):

        col = self.column_box.get()

        plt.plot(self.df[col])

        plt.title("Line Chart")

        plt.show()


    def bar_chart(self):

        col = self.column_box.get()

        plt.bar(range(len(self.df[col])), self.df[col])

        plt.title("Bar Chart")

        plt.show()


    def hist_chart(self):

        col = self.column_box.get()

        plt.hist(self.df[col])

        plt.title("Histogram")

        plt.show()



root = tk.Tk()

app = DataApp(root)

root.mainloop()
```
